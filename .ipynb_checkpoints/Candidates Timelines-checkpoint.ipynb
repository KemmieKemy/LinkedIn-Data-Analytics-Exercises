```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import os

plt.style.use('ggplot')
```

```python
#Create a folder for the data if it does not exist
try:
    os.makedirs('timelines')
    print("We just made a 'timelines' folder as it didn't exist before")
except:
    print(f"Folder already exists")
```

```python
#Read in the required file
atiku = pd.read_csv('Atiku_InFocus.csv')
atiku.shape
```

```python
#Peep the head.
atiku.head(2)
```

```python
#Convert created_at column to datetime
atiku['created_at'] = pd.to_datetime(atiku.created_at)
```

```python
#Get df of tweets within the period in focus
atiku = atiku[(atiku["created_at"] > '2018-10-31') & (atiku["created_at"] < '2019-02-28')]
```

```python
#A quick plot of distribution of tweets this period in 
atiku.created_at.plot(figsize=(15, 5))
```

```python
#Create a 'dates' column to hold year+month+day from created_at col
atiku['dates'] = [line.date() for line in atiku.created_at]

#Create a 'times' column to holding time (hour) of the day from created_at col
atiku['times'] = [line.time().hour for line in atiku.created_at]
```

```python
atiku.created_at[567].time().strftime('%I') 
```

```python
ax = atiku['times'].value_counts().head(10).plot(kind='bar', figsize=(20,5),
                                        grid=True,
                                        title='Top 10 Active Hours of Candidate Atiku')
ax.set_xlabel("Tweet Time")
ax.set_ylabel("No of Tweets")
ax.get_figure().savefig("timelines\Atiku_Top10_Hours_line._sorted_bar.png", bbox_inches="tight")
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
```

```python
ax = atiku['dates'].value_counts().head(10).plot(kind='bar', figsize=(20,5),
                                        grid=True,
                                        title='Top 10 Active Days of Candidate Atiku')
ax.set_xlabel("Tweet Date")
ax.set_ylabel("No of Tweets")
ax.get_figure().savefig("Atiku_Top10_Day_line._sorted_bar.png", bbox_inches="tight")
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
```

```python
ax = atiku['dates'].value_counts().plot(figsize=(20,5),
                                        grid=True,
                                        rot=80,
                                        title='Tweet Timeline of Candidate Atiku')
# ax = df2.plot(lw=2, , marker='.', markersize=10, )
ax.set_xlabel("Tweet Date")
ax.set_ylabel("No of Tweets")
ax.xaxis.set_major_locator(mdates.WeekdayLocator())
ax.get_figure().savefig("Atiku_Timeline_line.png", bbox_inches="tight")
```
